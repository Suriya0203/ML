```python
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('/content/drive/MyDrive/image_recognition/image[1].jpg')
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#plt.subplot(1,1,1)
plt.imshow(img_gray)
plt.show()
```

```python
stop_data=cv2.CascadeClassifier('/content/drive/MyDrive/datasets/stop_data[1].xml')
found=stop_data.detectMultiScale(img_gray,minSize=(20,20))
amount_found=len(found)
print(found)
```

```python
if amount_found!=0:
  for (x,y,width,height) in found:
    image=cv2.rectangle(img_rgb,(x,y),(x+height,y+width),(0,255,0) ,5)
print(img)
plt.imshow(image)
plt.show()
```
